```python
import pandas as pd
```

```python
df = pd.read_csv('csvFiles/weather_data.csv', parse_dates=['day'])
df
```

```python
df.set_index('day', inplace=True)
df
```

```python
# replace all null values
df.fillna(0)
```

```python
df
```

```python
df.temperature.mean()
```

```python
# filling the null spaces
df.fillna({
    'temperature': df.temperature.mean(),
    'windspeed': df.windspeed.mean(),
    'event': 'No event'
})
```

```python
df
```

```python
df.interpolate()
```

```python
df
```

```python
# drop all rows that have a NaN value
df.dropna()
```

```python
# only drop the rows that are all filled with NaN
df.dropna(how='all') # how='any'
```

```python
# at least 2 NaN
df.dropna(thresh=2) 
```
