```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
df
```

```python
df = pd.concat([india_weather, us_weather], ignore_index=True) # index in sequence
df
```

```python
df = pd.concat([india_weather, us_weather], keys=['india', 'us']) # separate in topics
df
```

```python
df.loc['india']
```

```python
df.loc['us']
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
# merge
df3 = pd.merge(df1, df2, on="city")
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
pd.merge(df1,df2,on="city", how='inner') # only common values - intersection
```

```python
# all the values in the table 1 (left) and intersection, but not the values in table 2 (right) without common values (like san diego)
pd.merge(df1,df2,on="city", how='left')
```

```python
# all the values in the table 2 (right) and intersection, but not the values in table 1 (left) without common values (like orlando and baltimore)
pd.merge(df1,df2,on="city", how='right')
```

```python
# everything, like a full join
pd.merge(df1,df2,on="city", how='outer')
```
