```python
import pandas as pd

df_sales = pd.read_excel('linechart.xlsx')
df_sales.head()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12, 4))
plt.plot(df_sales['Quarter'], df_sales['Fridge'], color='red', label='Fridge')
plt.plot(df_sales['Quarter'], df_sales['Dishwasher'], color='pink', label='Dishwasher')
plt.plot(df_sales['Quarter'], df_sales['Washing Machine'], color='purple', label='Washing Machine')

plt.title('Product Sales')
plt.ylabel('Revenue')
plt.xlabel('Financial Quarter')
plt.legend()

plt.show()
```

```python
total_sales = df_sales[['Fridge', 'Dishwasher', 'Washing Machine']].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct='%1.2f%%', explode=(0.1,0,0), shadow=True)
plt.show()
```

```python
df_sales.plot(kind='bar', x='Quarter')
plt.xticks(rotation=45)

plt.show()
```
